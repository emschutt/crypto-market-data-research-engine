# Data Quality Checks

This notebook inspects the sample Binance market-data capture and checks for expected columns, non-empty datasets, timestamp ordering, and simple null rates.

```python
from pathlib import Path
import sys
import pandas as pd

ROOT = Path.cwd()
if not (ROOT / 'src').exists():
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.storage import read_dataset

DATA_ROOT = ROOT / 'sample_data' / 'smoke'
trades = read_dataset(DATA_ROOT, 'raw_agg_trades')
depth = read_dataset(DATA_ROOT, 'raw_depth')
features = read_dataset(DATA_ROOT, 'features')

for name, df in {'trades': trades, 'depth': depth, 'features': features}.items():
    print(name, df.shape)
    display(df.head())
```

```python
checks = {
    'trades_non_empty': len(trades) > 0,
    'depth_non_empty': len(depth) > 0,
    'features_non_empty': len(features) > 0,
    'trade_columns': {'event_ts', 'trade_id', 'price', 'quantity', 'trade_side'}.issubset(trades.columns),
    'feature_columns': {'bar_ts', 'midprice', 'spread', 'order_flow_imbalance'}.issubset(features.columns),
}
checks
```

```python
null_rates = features.isna().mean().sort_values(ascending=False)
null_rates[null_rates > 0]
```
